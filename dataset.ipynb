## Identificação de valores nulos

Antes de realizar qualquer análise, é importante verificar a existência de dados faltantes.

Esta etapa contabiliza a quantidade de valores nulos em cada coluna, permitindo identificar variáveis que podem necessitar de tratamento ou limpeza.

```python
df.isnull().sum()
```

## Distribuição dos participantes por gênero

## Frequência dos níveis de depressão

Esta análise contabiliza quantos participantes pertencem a cada categoria de nível de depressão.

```python
df["Depression"].value_counts()
```

## Remoção de registros duplicados

Nesta etapa são identificadas e removidas linhas duplicadas do conjunto de dados.

A remoção de duplicidades evita que um mesmo registro seja contabilizado mais de uma vez, garantindo maior confiabilidade nos resultados das análises.

```python
print("Antes:", df.shape)

df = df.drop_duplicates()

print("Depois:", df.shape)
```

## Limpeza de registros incompletos

Após identificar a existência de valores ausentes, os registros contendo informações faltantes são removidos.

Essa etapa garante que as análises posteriores sejam realizadas apenas com dados completos.

```python
df = df.dropna()
```

## Validação da limpeza dos dados
